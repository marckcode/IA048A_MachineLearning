**Air Traffic Analysis: Trends, Seasonality, and Efficiency**

**Overview**
This comprehensive analysis of air traffic data explores various facets of the industry, including annual and monthly passenger volumes, load factors, and the relationship between the number of flights and passengers. The visualizations below, created with Power BI, provide clear insights into the trends and dynamics of air travel.


**Annual Passenger Trends**


The bar chart above demonstrates a general increase in air travel passengers over the years, with a notable peak around 2015. Thereafter, a slight decline is observed, culminating in a significant drop in 2020, which can be attributed to the COVID-19 pandemic's impact on global travel.


**Monthly Passenger Volume**

The monthly distribution of passengers indicates pronounced seasonality in air travel, with summer months (July and August) exhibiting the highest volumes. This is consistent with peak vacation periods. In contrast, January and February show the lowest passenger numbers, possibly due to post-holiday seasonality and less favorable weather conditions.


**Domestic vs. International Traffic**



When comparing domestic and international travel, domestic flights consistently show higher passenger volumes throughout the year. Both sectors peak during the summer, with a smaller peak in December, indicating holiday-related travel.


**Load Factor Trends**



The line graph reveals a steady increase in the average Load Factor (LF), suggesting improved efficiency in filling seats. However, a sharp decline around 2020 reflects the adverse effects of the pandemic, followed by a rebound, indicating a recovery phase or adjustments in capacity.


**Flight and Passenger Correlation**



Lastly, the scatter plot with a fitted trend line illustrates a positive correlation between the number of flights and passengers. This relationship suggests that an increase in flights generally leads to an increase in air travel passengers.
